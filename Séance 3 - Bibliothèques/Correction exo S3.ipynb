# Exercice 1

Récupérer une page Wikipédia (celle sur la suisse) avec la bibliothèque Requests et l'enregistrer dans un fichier texte

```python
# 1. j'importe requests
import requests

# 2. je fais une requête sur la page wikipedia de la suisse
suisse = requests.get("https://fr.wikipedia.org/wiki/Suisse")

# 3. et je stocke le contenu text
contenu_texte = suisse.text
```

```python
# vérifions si on a bien récupéré ce que l'on veut
contenu_texte
```

```python
# 4. j'utilise open en écriture -> "w" pour créer un fichier wiki_suisse
with open("wiki_suisse.txt", "w") as fichier:
    # 5. je demande d'écrire le contenu_texte dans le fichier créé
    fichier.write(contenu_texte)

# un fichier a bien été créé dans mon répertoire.
```

```python
# si jamais je veux vérifier :

# j'ouvre le fichier en question en mode lecture -> "r" puis j'affiche son contenu
with open("wiki_suisse.txt","r") as fichier:
    contenu_fichier = fichier.read()
    
contenu_fichier
```

Utiliser la bibliothèque permettant d'interagir avec wikipédia pour récupérer toutes les pages associées à la page de la Suisse. Regardez les fonctions disponibles dans la bibliothèque ...

### Option 1 avec wikipedia

```python
# importer la bibliothèque wikipedia
import wikipedia

# je configure la langue (ajout suite à la discussion lors de la séance)
wikipedia.set_lang("fr")

# je récupère la page de la suisse
page_suisse_wiki = wikipedia.page("Suisse")
page_suisse_wiki
```

#### En regardant la documentation et je trouve ça :
**links** : List of titles of Wikipedia page links on a page.\
**references** : List of URLs of external links on a page. May include external links within page that aren’t technically cited anywhere.

```python
# links à donc l'air pas mal
page_suisse_wiki.links
```

```python
# si on souhaite avoir également les pages externes à wikipedia il faudra utiliser .references
url_associees = page_suisse_wiki.references
url_associees
```

### Option 2 avec wikipediaapi

```python
# importer Wikipedia-API : wikipediaapi
import wikipediaapi
```

```python
# Extrait documentation :
# Getting single page is straightforward.
# You have to initialize Wikipedia object and ask for page by its name.
# It’s parameter language has be one of supported languages.

# je configure en langue française
wiki_wiki = wikipediaapi.Wikipedia('fr')
# et je récupère la page de la Suisse
page_suisse = wiki_wiki.page("Suisse")
```

```python
# de manière similaire, je peux récupérer les pages liées avec links :
page_suisse.links
```

```python
# En complément, la documentation donne un exemple de fonction que l'on peut adapter:
def print_links(page):
        links = page.links
        for title in sorted(links.keys()):
            print("%s: %s" % (title, links[title]))
```

```python
print_links(page_suisse)
```

```python
# Et que l'on peut généraliser pour l'utiliser plus facilement:
def print_links_cible(cible):
    wiki_wiki = wikipediaapi.Wikipedia('fr')
    page = wiki_wiki.page(cible)
    links = page.links
    for title in sorted(links.keys()):
        print("%s: %s" % (title, links[title]))
```

```python
print_links_cible("France")
```
